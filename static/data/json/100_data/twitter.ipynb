# **LIBRARY**

```python
import pandas as pd
import random
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# **READ DATA**

```python
df = pd.read_json('twitter.json')
df.head()
```

# **LABELLING & SPLITTING DATA**

```python
random.seed(45)
sentiment = [random.randint(0, 1) for i in range(100)]

df['sentiment'] = sentiment
df.sentiment.value_counts()
```

```python
df_train = df.drop(df.sample(5, random_state=40).index)
df_test = df.sample(5, random_state=40)
df_test
```

```python
df_train.tweet_text
```

# **TFIDF**

```python
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(df_train.tweet_text)

words_df = pd.DataFrame(vectors.toarray(), columns=vectorizer.get_feature_names())
words_df.head()
```

```python
X = words_df
y = df_train.sentiment
```

# **MODELLING**

```python
%%time

dtc = DecisionTreeClassifier()
dtc.fit(X, y)
```

```python
%%time

knn = KNeighborsClassifier()
knn.fit(X, y)
```

# **TEST**

```python
unknown_vectors = vectorizer.transform(df_test.tweet_text)
unknown_words_df = pd.DataFrame(unknown_vectors.toarray(), columns=vectorizer.get_feature_names())
unknown_words_df.head()
```

```python
unknown_words_df.shape
```

```python
df_test['pred_dtc'] = dtc.predict(unknown_words_df)
df_test['proba_dtc'] = dtc.predict_proba(unknown_words_df)[:, 1]

df_test['pred_knn'] = knn.predict(unknown_words_df)
df_test['proba_knn'] = knn.predict_proba(unknown_words_df)[:, 1]
```
